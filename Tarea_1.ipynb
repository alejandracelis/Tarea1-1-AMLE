Ejemplo: Existe una variable denominada `Coastline (coast/area ratio)`. Coastline es una medida de la cantidad de costa (acceso a mar) del país normalizada al área total del país para no beneficiar a países más grandes pero con la misma proporción de costa. A mayor "Costline" aumenta la capacidad portuaria per capita del país, más puertos facilita el comercio y podría aumentar el GDP per cápita.




---


**Population:** Representa la cantidad de habitantes del país. Un mayor número de
personas implica mayor consumo, lo que podría incrementar el PIB.

**Area:** El área total del país no tiene una relación directa con el PIB per cápita, ya que este depende más de la producción de bienes y servicios.

**Pop. Density:** Mide la cantidad de habitantes por milla cuadrada. Países más densamente poblados pueden tener mayor consumo, lo que influye en el PIB.

**Coastline (coast/area ratio):** Indica la cantidad de costa en relación con el área total. Una mayor costa puede facilitar el comercio portuario y contribuir al PIB.

**Net migration:** Refleja el saldo migratorio del país. Un saldo positivo puede estimular la economía al aumentar la fuerza laboral y el consumo.

**Infant mortality (per 1000 births):** Mide la cantidad de muertes infantiles por cada 1,000 nacimientos. Está negativamente correlacionado con el desarrollo y, por ende, con el PIB.

**Literacy (%):** Representa el porcentaje de la población alfabetizada. Una mayor tasa de alfabetización suele estar asociada con un mayor desarrollo y PIB.

**Phones (per 1000):** Indica la cantidad de teléfonos por cada 1,000 habitantes. Un mayor acceso a telecomunicaciones suele reflejar un mayor desarrollo económico y, por lo tanto, un PIB más alto.

**Arable (%):** Muestra el porcentaje del territorio apto para cultivos temporales. Un alto porcentaje de tierra cultivable suele estar relacionado con economías menos diversificadas y un menor PIB per cápita.

**Crops (%):** Indica la proporción de territorio destinada a cultivos permanentes. Al igual que la variable "Arable", puede estar relacionada con un menor PIB si la economía depende demasiado de la agricultura.

**Other (%):** Representa el porcentaje del territorio no dedicado a la agricultura. Países con economías desarrolladas suelen tener más territorio destinado a actividades industriales o de servicios, lo que puede correlacionarse con un mayor PIB.

**Climate:** Clasifica el tipo de clima predominante en el país. No tiene una relación directa con el PIB per cápita.

**Birthrate:** Mide el número de nacimientos por cada 1,000 habitantes. En países desarrollados, las tasas de natalidad tienden a ser bajas, lo que puede estar correlacionado con un alto PIB per cápita.

**Deathrate:** Indica el número de muertes por cada 1,000 habitantes. Altas tasas de mortalidad pueden estar relacionadas con menor desarrollo y un PIB más bajo.

**Agriculture:** Proporción del PIB proveniente del sector agrícola. A medida que los países se desarrollan, este porcentaje tiende a disminuir, por ende, podría tener una relación negativa con el PIB.

**Industry:** Indica el porcentaje del PIB generado por el sector industrial. En países en desarrollo, la industria crece hasta alcanzar un peak antes de dar paso a los servicios.

**Service:** Representa la proporción del PIB derivada del sector servicios. En países desarrollados, este sector domina la economía y se asocia con un alto PIB per cápita.

### Pregunta 1.15

En los ejemplos anteriores calculamos correlaciones para `GDP ($ per capita)` y `GDP ($)`. Genere un nuevo dataframe que tenga le variación porcentual de la correlación absoluta para cada una de las columnas de características, e.g., si la correlación en valor absoluto de `GDP ($ per capita)` vs `Industry` es 0.1 y la correlación `GDP ($)` vs `Industry` es 0.5, la variación deberá ser +500%. Dicha variación porcentual puede ser positiva o negativa, pero ordene los el dataframe de tal manera que la variación de correlación absoluta sea desendiente.

```python
variation_percentage = ((correlation_gdp_excluded - correlation_gdp) / correlation_gdp) * 100
```

```python
variation_df = pd.DataFrame({
    'Feature': correlation_gdp.index,
    'Variation (%)': variation_percentage
})

# Ordenar el DataFrame de manera descendente por la variación porcentual
variation_df = variation_df.sort_values(by='Variation (%)', ascending=False)

# Muestra el resultado
print(variation_df)
```

Para el año 2007, por cada país realice un ranking de las fuentes con más emisiones de CO2 excluyendo las variables `Per Capita` y `Bunker fuels (Not in Total)`. Es decir, asigne un número de 1 a 5 a $\{$ `Solid Fuel`, `Liquid Fuel`, `Gas Fuel`, `Cement`, `Gas Flaring` $\}$, donde 1 es la mayor fuente de emisión de ese país en aquel año, y 5 indica que fue la menor; así para todos los países.

Si en 2007 no se reporta una fuente de emisión para un país, por ejemplo, si emisiones de `Gas Flaring` no se reportara, entonces asigne números de 1 a 4 a las fuentes restantes. Análogo para un menor número de datos.

Luego, por cada variable grafique un histograma de frecuencias del ranking que obtuvo la fuente emisión a lo largo de todos los países.

¿Cuál fue la fuente más contaminante en la mayoría de países en 2007?


```python
df_co2["Year"] = df_co2["Year"].dt.year  # Extrae solo el año como número
df_2007 = df_co2[df_co2["Year"] == 2007]  # Filtra para 2007
print(df_2007)
```

```python
# Seleccionar columnas relevantes
fuentes_co2 = ["Country", "Solid Fuel", "Liquid Fuel", "Gas Fuel", "Cement", "Gas Flaring"]
df_2007 = df_2007[fuentes_co2]
```

```python
# Función para asignar ranking dentro de cada país
def rank_sources(row):
    valores = row[["Solid Fuel", "Liquid Fuel", "Gas Fuel", "Cement", "Gas Flaring"]].dropna()
    ranking = valores.rank(ascending=False, method="dense").astype(int)  # Ranking donde 1 es la mayor emisión
    return ranking
```

```python
# Función para asignar el ranking a las fuentes de emisión
def rank_sources(row):
    # Filtrar las fuentes reportadas (no NaN)
    valores = row[["Solid Fuel", "Liquid Fuel", "Gas Fuel", "Cement", "Gas Flaring"]].dropna()

    # Si hay fuentes reportadas, asignar un ranking
    if len(valores) > 0:
        ranking = valores.rank(ascending=False, method="dense").astype(int)  # Ranking donde 1 es la mayor emisión
    else:
        ranking = pd.Series([np.nan] * 5, index=["Solid Fuel", "Liquid Fuel", "Gas Fuel", "Cement", "Gas Flaring"])  # Si no hay fuentes, asignar NaN
    return ranking

# Aplicar el ranking a cada país
df_rankings = df_2007.set_index("Country").apply(rank_sources, axis=1)

# Resetear el índice para que "Country" sea una columna
df_rankings = df_rankings.reset_index()

# Mostrar el ranking de fuentes de CO2 para cada país en 2007
print(df_rankings)
```

```python
# Graficar histogramas de los rankings para cada fuente de emisión
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

# Crear histogramas
for i, source in enumerate(["Solid Fuel", "Liquid Fuel", "Gas Fuel", "Cement", "Gas Flaring"]):
    axes[i].hist(df_rankings[source].dropna(), bins=np.arange(0.5, 6.5, 1), edgecolor='black')
    axes[i].set_title(f'Histograma de Ranking para {source}')
    axes[i].set_xlabel('Ranking')
    axes[i].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```



---


Considerando la frecuencia, aproximadamente 160 países tenían "Liquid Fuel" en el top 1 en el año 2007. Por lo tanto, esta fuente fue la más contaminante en ese año.

Calcule el promedio a lo largo de toda la muestra ($\mathbb{E}[\cdot]$) para el cambio porcentual de cada país y genere una nueva serie con la resta entre el cambio porcentual del país $i$ en el año $t$, y el promedio del cambio porcentual del país $i$. En otras palabras, genere una serie con _**desvíos del cambio porcentual promedio**_ $\forall i,t$:

```python
paises = ["United Kingdom", "Canada", "Germany", "France", "United States", "Japan", "Italy", "Spain"]
start1 = pd.to_datetime("2007")
end1 = pd.to_datetime("2010")
start2 = pd.to_datetime("2017")
end2 = pd.to_datetime("2020")

# Filtrar por países y años (en datetime)
df_periodo1 = df_co2[(df_co2["Country"].isin(paises)) & (df_co2["Year"].between(start1, end1))]
df_periodo2 = df_co2[(df_co2["Country"].isin(paises)) & (df_co2["Year"].between(start2, end2))]
```
